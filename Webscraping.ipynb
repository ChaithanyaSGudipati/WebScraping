The aim of this project is to scrape data using Beautiful soup and using NLTK, plotting frequency distribution of novel from 'https://www.gutenberg.org/files/2701/2701-h/2701-h.htm'



## Importing required libraries

```python
import nltk
nltk.download('stopwords')
```

```python
import requests
from bs4 import BeautifulSoup
from nltk.tokenize import RegexpTokenizer
import nltk
import matplotlib.pyplot as plt
import seaborn as sns


r = requests.get('https://www.gutenberg.org/files/2701/2701-h/2701-h.htm')
html = r.text
soup = BeautifulSoup(html, "html5lib") # Creating beautiful soup object
text = soup.get_text() # Getting text from object
tokenizer = RegexpTokenizer('\w+') # Creating tokenizer and tokens
tokens = tokenizer.tokenize(text)
words = []
for word in tokens:
	words.append(word.lower())
	

sw = nltk.corpus.stopwords.words('english') # To get stopwords from nltk corpus
words_ns = []

for word in words:
	if word not in sw:
		words_ns.append(word)

# Frequency distribution plot
freqdist1 = nltk.FreqDist(words_ns)
freqdist1.plot(25)
```
